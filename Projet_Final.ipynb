## Scrap Image BeautifulSoup/Selenium

###  Scrap Image API

```python
# importing required libraries
import requests
from bs4 import BeautifulSoup

# target URL
url = "https://royaleapi.com/cards/popular?time=7d&mode=grid&cat=TopLadder&sort=rating"

headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    }

response = requests.request("GET", url, headers=headers)

data = BeautifulSoup(response.text, 'html.parser')

# find all with the image tag
images = data.find_all('img', src=True)

# select src tag
image_src = [x['src'] for x in images]

# select format images
image_src = [x for x in image_src if x.startswith('https://cdn.royaleapi.com/static/img/cards-150/')]

print('Number of Images: ', len(image_src))

image_count = 1
for image in image_src:
    with open('img2/image_'+str(image_count)+'.jpg', 'wb') as f:
        res = requests.get(image)
        f.write(res.content)
    image_count = image_count+1
if image_count == len(image_src): 
        print("All Images Downloaded!") 
else: 
    print(f"Total {image_count-1} Images Downloaded Out of {len(image_src)}") 
```

### Scrap Image WebSite

```python
def folder_create(images): 
    try: 
        folder_name = 'img' 
        
        os.mkdir(folder_name)
    except: 
        print("Folder Exist with that name!") 
        folder_create() 
    download_images(images, folder_name) 
  
  
def download_images(images, folder_name): 
    count = 0
    print(f"Total {len(images)} Image Found!") 
    if len(images) != 0: 
        for i, image in enumerate(images): 
            try: 
                image_link = image["data-srcset"] 
            except: 
                try: 
                    image_link = image["data-src"] 
                except: 
                    try: 
                        image_link = image["data-fallback-src"] 
                    except: 
                        try: 
                            image_link = image["src"]   
                        except: 
                            pass
            try: 
                r = requests.get(image_link).content 
                try: 
                    r = str(r, 'utf-8') 
                except UnicodeDecodeError: 
                    with open(f"{folder_name}/images{i+1}.jpg", "wb+") as f: 
                        f.write(r) 
                    count += 1
            except: 
                pass
        if count == len(images): 
            print("All Images Downloaded!") 
        else: 
            print(f"Total {count} Images Downloaded Out of {len(images)}") 
  
def main(url): 
    
    r = requests.get(url) 
    soup = BeautifulSoup(r.text, 'html.parser') 
    images = soup.findAll('img') 
    folder_create(images) 
 
# url = input("Enter URL:- ") 

url = 'https://statsroyale.com/fr/top/cards'

main(url)
```

```python
ListModeleDF = []
ListModeleCards = []
for articles in range(1,107):
    rank = browser2.find_element_by_xpath("/html/body/div[3]/div[4]/div[2]/div[3]/div[3]/div[2]/div["+str(articles)+"]/div[1]")
    notation = browser.find_element_by_xpath("/html/body/div[3]/div[4]/div[2]/div[3]/div[3]/table/tbody/tr["+str(articles)+"]/td[3]")
    utilisation = browser.find_element_by_xpath("/html/body/div[3]/div[4]/div[2]/div[3]/div[3]/table/tbody/tr["+str(articles)+"]/td[4]")
    win = browser.find_element_by_xpath("/html/body/div[3]/div[4]/div[2]/div[3]/div[3]/table/tbody/tr["+str(articles)+"]/td[6]")
    name = browser.find_element_by_xpath("/html/body/div[3]/div[4]/div[2]/div[3]/div[3]/table/tbody/tr["+str(articles)+"]/td[2]")
    
    ListModeleCards.append(str(name.text))
    ListModeleCards.append(str(rank.text))
    ListModeleCards.append(str(notation.text))
    ListModeleCards.append(str(utilisation.text))
    ListModeleCards.append(str(win.text))
    ListModeleDF.append(ListModeleCards[:])
    ListModeleCards.clear()
    print(rank.text,notation.text,utilisation.text,win.text,name.text)

browser.close()
browser2.close()
```

#### Create DataFrame

```python
df = pd.DataFrame(ListModeleDF)    
df.columns += 1 
df.to_csv("clash_stats.csv")
```
